# Adding additional variables

We are using this notebook to two additional variables: <br>
<ul>
    <li>Grade Coefficient</li>
    <li>Semester after registration</li>
</ul>
<br>
The first represents the Grade Coefficient used in my University.<br>
Each letter grade (A, A-, B+ etc) has a corresponding GPA value.<br><br>
Semester after registration represents the number of semesters that have passed since the student's registration.<br> If a student is fourth year second semester, the number of semesters after registration will be eight

<br><br> Importing standard packages

```python
import pandas as pd
```

Reading the data

```python
true_data = pd.read_csv('final_true_data.csv')
coeff = pd.read_csv('coeff.csv')
```

```python
coeff
```

```python
true_data
```

Adding the <b>Grade Coefficient</b> with merge

```python
df_with_coeff = pd.merge(true_data, coeff, how='left', left_on='Letter Grade', right_on='Num Grade')
```

```python
df_with_coeff.rename(columns={'Num Grade_x':'Num Grade'}, inplace=True)
```

```python
df_with_coeff.drop(['Num Grade_y', 'Unnamed: 0', 'Unnamed: 0.1'], inplace=True, axis=1)
```

```python
df_with_coeff
```

Adding the <b>Semester after registration</b>

```python
test_array = df_with_coeff['Semester'].unique()
```

```python
df_with_coeff['Semester'][df_with_coeff['Semester']=='7.Semester: Semester: 2020'] = '7.Semester: 2020 Spring'
```

```python
for semester in test_array:
    if semester != 'Unknown':
        print(semester)
```

```python
by_student = df_with_coeff.groupby('Student ID', as_index=False)
```

Function used to order correctly semesters and apply respective numeric values

```python
new_df = pd.DataFrame()
for student in by_student:
    student_df = student[1]
    student_df['Semester'] = student_df['Semester'].str.split(':',expand=True)[1]
    student_df['Semester'][student_df['Semester'].isna()] = 'Unknown'
    semesters = student_df['Semester'].unique()
    semesters = sorted(semesters, key=lambda semester: f"{semester[:5]}a"
                      if 'Spring' in semester else f"{semester[:5]}b")
    student_df['Semester after registration'] = student_df['Semester'].apply(lambda sem: semesters.index(sem)+1
                                                                            if 'Unknown' not in sem else 99)
    new_df=new_df.append(student_df, ignore_index = True)
```

The new data frame

```python
new_df
```

##### Adding two more variables
##### Whether a course is program or non-program 

```python
course_program_dict = {
    'Genetics and Bioengineering (Bachelor)':'BIO',
    'Architecture (Bachelor)': 'ARCH',
    'Computer Sciences and Engineering (Bachelor)':'CS',
    'Electrical and Electronics Engineering (Bachelor)':'EE',
    'Industrial Engineering (Bachelor)': 'IE',
    'Mechanical Engineering (Bachelor)' : 'ME',
    'Software Engineering (Bachelor)':'SE'
    
}
```

```python
fens_courses = ['BIO', 'ARCH', 'CS', 'EE', 'IE', 'ME', 'SE', 'MATH', 'ENS', 'NS']
```

```python
def f(row):
    if not isinstance(row['Course code'], float):
        if course_program_dict[row['Program']] in row['Course code']:
            return 'Program course'
        return 'Non-program course'
    return float('nan')
    
```

```python
def fac(row):
    if not isinstance(row['Course code'], float):
        if ''.join(i for i in row['Course code'] if not i.isdigit()) in fens_courses:
            return 'FENS course'
        return 'Non-FENS course'
    return float('nan')
    
```

```python
new_df['Program Course'] = new_df.apply(lambda x: f(x), axis=1)
```

```python
new_df['FENS Course'] = new_df.apply(lambda x: fac(x), axis=1)
```

```python
new_df.to_csv('data_added_variables.csv')
```
